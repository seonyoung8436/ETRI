```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("All/only_freq.csv")
test = pd.read_csv("Desktop/s/only_freq/4core/wrf.csv")
# 0: astar
# 1: calculix
# 2: gamess
# 3: gcc
# 4: gobmk
# 5: mcf
# 6: povray
# 7: soplex
# 8: sphinx3
# 9: tonto
# 10: wrf
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression(n_jobs=1)
lreg.fit(data.iloc[:,:-2], data.iloc[:,-1])
```

```python
preds = lreg.predict(test.iloc[:,:-2])
resultData = DataFrame(preds, columns = ['preds'])
resultData['temp'] = test.iloc[:,-1]
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test.iloc[:,-1],preds)
```

```python
resultData.plot(figsize=(40,15), grid=1, title="Temperature Prediction",ylim=[0,90])
plt.xlabel("Time(10ms)")
plt.ylabel("Temperature")
plt.show()
```
